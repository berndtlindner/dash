```python
# https://medium.com/@harimittapalli/exploratory-data-analysis-iris-dataset-9920ea439a3e
import pandas as pd
from sklearn.datasets import load_iris
import pandas_bokeh
import ipywidgets as widgets
from ipywidgets import interact, interactive, fixed, interact_manual
import plotly_express as px
import pandas_bokeh
import bokeh
from IPython.display import display, clear_output
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

pandas_bokeh.output_notebook()

dataset=load_iris()
data=pd.DataFrame(dataset['data'],columns=['Petal length','Petal Width','Sepal Length','Sepal Width'])
data['Species']=dataset['target']
data['Species']=data['Species'].apply(lambda x: dataset['target_names'][x])

select = widgets.SelectMultiple(
    options=list(data.Species.unique()),
    value=list(data.Species.unique()),
    description='Species',
    disabled=False
)

# display(select)
```

```python
@interact(x=select)
def g(x):
    user_data= data[data.Species.isin(x)]
    plot = px.scatter(
    user_data,
    x="Sepal Length",
    y="Sepal Width",
    color="Species",
    height=700
    )
#     p_scatter = user_data.plot_bokeh(
#         kind="scatter",
#         x="Petal length",
#         y="Petal Width",
#         category="Species",
#         title="Iris DataSet Visualization",
#         show_figure=False
#         )
    # pandas_bokeh.show(p_scatter)
    return display(plot, user_data, 
                   #pandas_bokeh.show(p_scatter)
                  )
```

# Only run user clicks/runs

```python
select2 = widgets.SelectMultiple(
    options=list(data.Species.unique()),
    value=list(data.Species.unique()),
    description='Species',
    disabled=False
)

button = widgets.Button(description="Transform & Display")
output = widgets.Output()

def on_button_clicked(b):
    with output:
        clear_output()
        user_data= data[data.Species.isin(select2.value)]
        plot = px.scatter(
        user_data,
        x="Sepal Length",
        y="Sepal Width",
        color="Species",
        height=700
        )
        display(user_data, plot)

display(select2, button)

button.on_click(on_button_clicked)
```
